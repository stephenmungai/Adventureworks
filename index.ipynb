* Notes 
1) For uniformity in sales we will only consider the sales in USD

```python
# Filtering out the dataset for purchases made in USA or transacted in USD currency
df_filtered = df[(df['country'] == 'United States') | (df['from_country'] == 'USD')]

# Check the shape of the filtered dataset
df_filtered.shape
```

```python
# Check the first 30 rows of the filtered dataset

df_filtered.head(30)
```

```python
df_filtered['from_country'].value_counts()
```

* Insights

1) We have filtered the dataset to include sales that were either made in United States or were originally in USD.

```python
df_filtered.info()
```

```python
# Lets drop the columns that contain null values

df_filtered =df_filtered.drop(columns= ['currency_rate','from_country','to_country'])
```

```python
# Check the descriptive statistics of the filtered dataset

df_filtered.describe()
```

```python
# Check value counts of the country column,product_name and product_category columns

column_names =['country','product_name','product_category']

for column in column_names:
    print(f"Value counts for {column}:")
    print(df_filtered[column].value_counts())
    print("\n")
```

```python
# Changing the data type of the date_of_purchase column to datetime

df_filtered['date_of_purchase'] = pd.to_datetime(df_filtered['date_of_purchase'], format='%Y-%m-%d %H:%M:%S.%f')

# Retrieve the Year and month from the date_of_purchase column and create new columns to store the values

df_filtered['year'] = df_filtered['date_of_purchase'].dt.year
df_filtered['month'] = df_filtered['date_of_purchase'].dt.strftime('%B')

# drop the date_of_purchase column

df_filtered = df_filtered.drop(columns=['date_of_purchase'])

# Check the first 10 rows of the filtered dataset
df_filtered.head(10)
```

```python
df_filtered['year'] = df_filtered['year'].astype(str)  # Convert year to string for better plotting
```

```python
# Check value counts for the year column

df_filtered['year'].value_counts()
```

```python
# Rounding Up the total_price column to 2 decimal places
df_filtered['total_price'] = df_filtered['total_price'].round(2)
# Check the first 10 rows of the filtered dataset
df_filtered.head(10)
```

```python
# Create a unit price column by dividing the total_price by quantity
df_filtered['unit_price'] = df_filtered['total_price'] / df_filtered['quantity']

# Round the unit price to 2 decimal places
df_filtered['unit_price'] = df_filtered['unit_price'].round(2)
# reorder the columns in the dataset
df_filtered = df_filtered[['first_name', 'last_name', 'country', 'product_name', 'product_category', 'quantity', 'unit_price', 'total_price', 'year', 'month']] 

# Check the first 10 rows of the filtered dataset
df_filtered.head(10)
```

```python
# Dropping the first_name and last_name columns as they are not needed for analysis

df_filtered = df_filtered.drop(columns=['first_name', 'last_name'])
```

```python
df_cleaned = df_filtered.copy()
```

# Data Visualization

```python
df_cleaned.head(10)
```

### Bar chart to visualise total sales by country


```python
df_total_sales_by_country = df_cleaned.groupby('country')['total_price'].sum().reset_index()

# Sort the total sales by country in descending order
df_total_sales_by_country = df_total_sales_by_country.sort_values(by='total_price', ascending=False)

# Plotting the total sales by country
plt.figure(figsize=(12, 6))
sns.barplot(x='country', y='total_price', data=df_total_sales_by_country, palette='viridis',hue='country')
plt.title('Total Sales by Country')
plt.xlabel('Country')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

* Insights

1) The bar chart shows that the highest sales are from the United States, followed by Canada and then the rest of the world.
2) Sales made in United States are significantly higher than sales made in other countries.

## Bar chart showing the total sales by product category

```python
df_total_sales_by_product_category = df_cleaned.groupby('product_category')['total_price'].sum().reset_index()
# Sort the total sales by product category in descending order
df_total_sales_by_product_category = df_total_sales_by_product_category.sort_values(by='total_price', ascending=False)
# Plotting the total sales by product category
plt.figure(figsize=(12, 6))
sns.barplot(x='product_category', y='total_price', data=df_total_sales_by_product_category, palette='viridis',hue='product_category')
plt.title('Total Sales by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show();
```

* Insights

1) The bar chart shows that the highest sales are from the sale of bikes
2) The second highest sales are sales from components

- We can therefore infer that the major source of sales is from the sale of bikes

## Bar chart showing the number of products in each product category

```python
df_product_by_product_category = df_cleaned.groupby('product_category')['product_name'].nunique().reset_index()

# Sort the number of unique products by product category in descending order

df_product_by_product_category = df_product_by_product_category.sort_values(by='product_name', ascending=False)
# Plotting the number of unique products by product category
plt.figure(figsize=(12, 6))
sns.barplot(x='product_category', y='product_name', data=df_product_by_product_category, palette='viridis',hue='product_category')
plt.title('Number of Unique Products by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Number of Unique Products')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show();
```

## Bar chart showing the 10 most bought products in each category

```python
df_product_and_quantity_by_product_category = df_cleaned.groupby(['product_category', 'product_name'])['quantity'].sum().reset_index()

#Sort the quantity by product category and product name in descending order
df_product_and_quantity_by_product_category = df_product_and_quantity_by_product_category.sort_values(by='quantity', ascending=False)

# Plotting several plots for each product category showing the 10 most sold products
for category in df_product_and_quantity_by_product_category['product_category'].unique():
    plt.figure(figsize=(12, 6))
    sns.barplot(x = 'product_name', y= 'quantity', data = df_product_and_quantity_by_product_category[df_product_and_quantity_by_product_category['product_category'] == category].head(10), palette='viridis',hue='product_name')
    plt.title(f'Top 10 Most Sold Products in {category} Category')
    plt.xlabel('Product Name')
    plt.ylabel('Quantity Sold')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show();

```

* Insights

1) Components has the highest number of unique products
2) The second highest number of unique products is in the bikes category
3) The components category contains spare parts for bikes for example bike frames and bike handle bars for the bikes sold by the company
4) Accesories contains additional items that are not necessary for the bike to function for example bike locks and bike helmets
5) The bikes category contains the actual bikes sold by the company


- As components category contains spare parts for bikes, this is expected as various spare parts are needed to build a bike.



## Histogram showing the distribution of bike sales

```python
fig = plt.figure(figsize=(12, 6))
# Plotting the distribution of unit price on bike category

sns.histplot(df_cleaned[df_cleaned['product_category'] == 'Bikes']['unit_price'], bins=30, kde=True, color='blue')
plt.title('Distribution of Unit Price for Bikes')
plt.xlabel('Unit Price')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

### Histogram showing distribution of unit_price of bikes over the years

```python
# PLots showing the distribution of unit price of bikes category by year

for year in df_cleaned['year'].unique():
    plt.figure(figsize=(12, 6))
    sns.histplot(df_cleaned[(df_cleaned['product_category'] == 'Bikes') & (df_cleaned['year'] == year)]['unit_price'], bins=30, kde=True, color='blue')
    plt.title(f'Distribution of Unit Price for Bikes in {year}')
    plt.xlabel('Unit Price')
    plt.ylabel('Frequency')
    plt.tight_layout()
    plt.show()
```

* Insights
1) There is no clear trend in the distribution of bike sales
2) There is no clear trend in the distribution of unit price of bikes over the years


- This could be due to the fact that the dataset is not large enough to show a clear trend in the distribution of bike sales and unit price of bikes over the years.
- The dataset also lacks the subcategories of the bikes sold by the company. For example, mountain bikes, road bikes and hybrid bikes. This could also be a reason for the lack of trend in the distribution of bike sales and unit price of bikes over the years.

## Line Graph showing the trend of sales over the years

```python
df_sales_over_years = df_cleaned.groupby(['year'])['total_price'].sum().reset_index()

# Sort the sales data by year and month
df_sales_over_years = df_sales_over_years.sort_values(by=['year'])
# Plot the total sales over the years
plt.figure(figsize=(12, 6))
sns.lineplot(x='year', y='total_price', data=df_sales_over_years, marker='o')
plt.title('Total Sales Over the Years')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
df_sales_over_years
```

### Line Graph showing the product caregories over the years 

```python
df_sales_categories_over_years = df_cleaned.groupby(['year', 'product_category'])['total_price'].sum().reset_index()

# Plot the total sales by product category over the years
plt.figure(figsize=(12, 6))
sns.lineplot(x='year', y='total_price', hue='product_category', data=df_sales_categories_over_years, marker='o')
plt.title('Total Sales by Product Category Over the Years')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.legend(title='Product Category')
plt.tight_layout()
plt.show()
```

### Line Graph showing trend of sales by year and month

```python
# Convert the month column to a categorical type with the correct order
df_cleaned['month'] = pd.Categorical(df_cleaned['month'], categories=['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December'], ordered=True)

# Group the data by month and year, and calculate the total sales
df_sales_by_month = df_cleaned.groupby(['year', 'month'],observed= True)['total_price'].sum().reset_index()
# Sort the sales data by year and month
df_sales_by_month = df_sales_by_month.sort_values(by=['year', 'month'])
# Plot the total sales by month and year
plt.figure(figsize=(12, 6))
sns.lineplot(x='month', y='total_price', hue='year', data=df_sales_by_month, marker='o')
plt.title('Total Sales by Month and Year')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.legend(title='Year')
plt.tight_layout()
plt.show();
```

### Line graph showing trend of sales by year and month for each product category

```python
df_sales_of_product_category_by_month = df_cleaned.groupby(['year', 'month', 'product_category'],observed= True)['total_price'].sum().reset_index()

# Sort the sales data by year, month, and product category

df_sales_of_product_category_by_month = df_sales_of_product_category_by_month.sort_values(by=['year', 'month', 'product_category'])
# Plot multiple plots for the different years

for year in df_sales_of_product_category_by_month['year'].unique():
    plt.figure(figsize=(12, 6))
    sns.lineplot(x='month', y='total_price', hue='product_category', data=df_sales_of_product_category_by_month[df_sales_of_product_category_by_month['year'] == year], marker='o')
    plt.title(f'Total Sales by Month and Product Category in {year}')
    plt.xlabel('Month')
    plt.ylabel('Total Sales')
    plt.xticks(rotation=45)
    plt.legend(title='Product Category')
    plt.tight_layout()
    plt.show();

```

* Insights 

1) The line graph shows that the sales have been increasing over the years
2) The dataset contains sales data from 2011 to 2014 with two incomplete years (2011 and 2014)
3) For the two incomplete years, the sales data is not complete for the months of January and December. This could be due to the fact that the dataset was collected over a period of time and the sales data for these months was not available at the time of collection or the source database being incomplete due to being a test database.
4) If we check sales data compared to to months of the year, we can see that the sales are highest in the months of May and June. This could be due to the fact that these months are summer months in the United States and Canada and people are more likely to buy bikes during these months.
5) The sales of clothing and accessories are consisten throughout the year. This could be due to the fact that these items are not seasonal and people are more likely to buy them throughout the year.

# Recommendations

1. **Focus on Bikes Sales**: 
    - Bikes contribute the highest revenue among all product categories. The company should prioritize marketing and inventory management for bikes to maximize sales.

2. **Expand Components Offerings**:
    - Components have the highest number of unique products and are essential for bike assembly and maintenance. Expanding the components product line could attract more customers and increase sales.

3. **Seasonal Promotions**:
    - Sales are highest during the summer months (May and June). The company should plan seasonal promotions and discounts during these months to capitalize on increased demand.

4. **Target United States Market**:
    - The United States accounts for the majority of sales. The company should continue to focus on this market while exploring opportunities to grow sales in other regions like Canada and Australia.

5. **Diversify Product Categories**:
    - While bikes dominate sales, other categories like accessories and clothing have consistent demand throughout the year. Investing in these categories could provide steady revenue streams.

6. **Leverage High-Selling Products**:
    - Products like "Mountain-100 Black" and "Sport-100 Helmet" are among the most sold. The company should ensure sufficient stock and promote these high-demand items.

7. **Explore International Markets**:
     - While the United States is the primary market, there is potential to grow sales in other countries like Canada, Australia, and the United Kingdom through targeted marketing campaigns.

8. **Enhance Customer Experience**:
     - Consider offering bundled deals (e.g., bikes with accessories) to encourage customers to purchase multiple items, thereby increasing the average order value.
